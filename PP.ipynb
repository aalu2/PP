```python
#Answer of 1
class Solution:
    def reverseWords(self, s: str) -> str:      
        #Splitting string on whitespaces 
        words = s.split() 
        #Reversing the words
        words = [word[::-1] for word in words]  
        #Joining the word in list
        words = ' '.join([word for word in words])  
        return words
    
    #Test    
    print(reverseWords("String","Hello Python"))
```

```python
#Answer of 2
def sum_indices(nums, target):
    for i in range(len(nums) - 1):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]
    return None

nums = [1,2,3,4]
target = 7

print(sum_indices(nums, target))
```

```python
#Answer of 3
class Solution:
    def isPalindrome(self, x: int) -> bool:
        tmp = x
        rev = 0

        #For negative integers
        if x < 0: 
          return False

        #For positive integers
        while tmp != 0:
          rev = ( rev*10 ) + ( tmp%10)
          tmp = tmp // 10

        if x == rev:
          return True
        else:
          return False

object1 = Solution ()
print( object1.isPalindrome(121) )
print( object1.isPalindrome(-121) )
print( object1.isPalindrome(100) )
```

```python
#Answer of 4
def longestCommonPrefix( a):   
    size = len(a)
 
    # size is 0, so returning empty string
    if (size == 0):
        return ""
    if (size == 1):
        return a[0]
 
    # sorting the array of strings
    a.sort()
     
    # finding the minimum length from
    # first and last string
    end = min(len(a[0]), len(a[size - 1]))
 
    # finding the common prefix between
    # the first and last string
    i = 0
    while (i < end and
           a[0][i] == a[size - 1][i]):
        i += 1
    pre = a[0][0: i]
    return pre
 
# Testing
if __name__ == "__main__":
 
    input = ["michael", "michelle", "mitch"] 
    print("The longest Common Prefix is :" , longestCommonPrefix(input))
```

```python
#Answer of 6
from math import *
def average(a, b):
    return (a + b) / 2.0
def improve(guess, x):
    return average(guess, x/guess)
def good_enough(guess, x):
    d = abs(guess*guess - x)
    return (d < 0.001)
def square_root(guess, x):
    while(not good_enough(guess, x)):
        guess = improve(guess, x)
    return guess
def my_sqrt(x):
    r = square_root(1, x)
    return r
#Test
print(my_sqrt(16))

#Testing accuracy
print(my_sqrt(16)-math.sqrt(16))
```
